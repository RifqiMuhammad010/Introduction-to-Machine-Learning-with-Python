# **Bagian 4: Siklus Umpan Balik dan Human-in-the-Loop**
Pemodelan data bukanlah garis lurus, melainkan sebuah siklus iteratif (feedback loop). Sering kali, hasil dari sebuah model mengungkap inkonsistensi pada anotasi data atau ketidakcukupan fitur prediktif. Kesalahan (error) dari model menjadi landasan utama untuk mendefinisikan ulang fitur, memodifikasi label, atau merestrukturisasi masalah itu sendiri.

Selain itu, sistem tidak selamanya diizinkan berjalan otonom secara absolut. Pada skenario dengan risiko tinggi (high-stakes decision), pendekatan Human-in-the-Loop diterapkan.

| Kondisi Implementasi | Landasan Penerapan |
| :--- | :--- |
| **Risiko Keputusan Ekstrem** | Diagnosis medis wajib divalidasi oleh dokter ahli. |

# **Bagian 5: Transisi dari Prototype ke Production**
Sebuah prototype bertujuan membuktikan bahwa data yang dimiliki memiliki sinyal kuat terhadap target prediksi (Proof of Concept). Namun, mendeploy model ini ke sistem aktual (Production System) membutuhkan arsitektur keandalan (reliability) kelas enterprise.

Eksperimen vs Keandalan: Prototype menekankan kecepatan eksplorasi, sementara produksi menekankan stabilitas infrastruktur (seperti penanganan data stream, persistensi memori, dan latensi query).

Technical Debt (Utang Teknis): Semakin kompleks sebuah rentetan pipeline (prapemrosesan bertumpuk, ensambel model raksasa), semakin sulit dan mahal perawatannya saat distribusi data pengguna berubah (Data Drift). Kesederhanaan (Occam's Razor) diutamakan dalam produksi.

# **Bagian 6: Metodologi Pengujian A/B (A/B Testing)**
Skor akurasi pada himpunan data uji statis (pengujian offline) tidak menjamin kinerja yang ekuivalen pada sistem waktu-nyata. Pengujian daring (online testing) yang paling empiris dan terpercaya adalah A/B Testing.

Pendekatan ini membagi pengguna aktif secara acak ke dalam dua lingkungan operasi:

Grup A (Kontrol): Menggunakan sistem model lama atau sistem baseline.

Grup B (Varian): Menggunakan model rekomendasi prediktif yang baru.

```
# Simulasi Analisis Laporan A/B Testing Konversi Pengguna
ab_test = pd.DataFrame({
    "Kelompok Uji": ["Grup A (Sistem Lama)", "Grup B (Model Rekomendasi Baru)"],
    "Total Pengguna": [10000, 10000],
    "Jumlah Konversi (Beli)": [820, 910]
})

# Perhitungan Tingkat Konversi
ab_test["Conversion Rate (%)"] = (ab_test["Jumlah Konversi (Beli)"] / ab_test["Total Pengguna"]) * 100
display(ab_test)
```

# **Bagian 7: Rekayasa Estimator Kustom (Custom Estimators)**
Terkadang, langkah prapemrosesan khusus yang bergantung pada arsitektur data lokal tidak tersedia secara bawaan di scikit-learn. Agar manipulasi manual tersebut tetap aman dari kebocoran data dan selaras dengan arsitektur Pipeline maupun fungsi evaluasi cross_val_score, kita wajib membungkus proses tersebut ke dalam kelas transformer buatan sendiri. Hal ini dilakukan dengan mewarisi atribut dari BaseEstimator dan TransformerMixin.

```
# 1. Definisi Custom Transformer menggunakan Paradigma Pemrograman Berorientasi Objek
class AddConstantFeature(BaseEstimator, TransformerMixin):
    """
    Transformer buatan khusus untuk menambahkan satu kolom bernilai konstan
    ke matriks fitur secara otomatis di dalam pipeline.
    """
    def __init__(self, value=1.0):
        self.value = value

    def fit(self, X, y=None):
        # Fit tidak melakukan kalkulasi apapun, hanya mengembalikan instans objek
        return self

    def transform(self, X):
        # Operasi matriks numpy untuk menyematkan kolom tambahan
        constant_column = np.full((X.shape[0], 1), self.value)
        return np.hstack([X, constant_column])

# 2. Persiapan Data Latih
cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(
    cancer.data, cancer.target, random_state=42, stratify=cancer.target
)

# 3. Integrasi Transformer Kustom ke dalam Pipeline Scikit-Learn Resmi
pipe_custom = Pipeline([
    ("tambah_kolom_konstan", AddConstantFeature(value=1.0)), # Memanggil class buatan sendiri
    ("scaler_standar", StandardScaler()),
    ("model_klasifikasi", LogisticRegression(max_iter=5000))
])

# Eksekusi Cross Validation dengan Pipeline Custom
scores = cross_val_score(pipe_custom, cancer.data, cancer.target, cv=5)
print(f"Skor CV Rata-Rata menggunakan Transformer Kustom: {scores.mean():.3f} (Std: {scores.std():.3f})")
```
